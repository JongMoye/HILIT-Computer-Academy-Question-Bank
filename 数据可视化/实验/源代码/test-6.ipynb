```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus'] = False
```

```python

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot([0, 1], [0, 1])

# 标注坐标轴组件
ax.annotate('XAxis (axis.XAxis)', xy=(0.5, -0.1), xytext=(0.5, -0.2), ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate('YAxis (axis.YAxis)', xy=(-0.1, 0.5), xytext=(-0.2, 0.5), va='center', arrowprops=dict(arrowstyle="->"))
ax.annotate('Spine (spines.Spine)', xy=(0, 0), xytext=(0.2, 0.2), arrowprops=dict(arrowstyle="->"))
ax.annotate('Major Ticks', xy=(0.5, 0), xytext=(0.5, 0.1), ha='center', arrowprops=dict(arrowstyle="->"))
ax.annotate('Minor Ticks', xy=(0.25, 0), xytext=(0.25, 0.05), ha='center', arrowprops=dict(arrowstyle="->"))

plt.title("Matplotlib坐标轴组件结构", pad=20)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
# 主坐标轴（默认）
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(np.random.rand(10))
ax1.set_title('Main Axes')

# 自定义位置的小坐标轴
ax2 = fig.add_axes([0.6, 0.6, 0.2, 0.2])
ax2.plot(np.random.rand(5), color='red')
ax2.set_title('Custom Axes')
```

```python
fig, ax = plt.subplots()
x = np.arange(0, 10, 0.1)
ax.plot(x, np.sin(x))

# 设置主刻度为π/2间隔
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi/2))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: f'{val/np.pi:.0f}π'))

# 设置次刻度并调整样式
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi/4))
ax.tick_params(axis='x', which='minor', length=4, color='r')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# 隐藏上/右轴脊
ax1.plot(np.random.rand(10))
ax1.spines[['top', 'right']].set_visible(False)

# 移动轴脊到数据零点
ax2.plot(np.arange(-5, 5), np.arange(-5, 5))
ax2.spines['left'].set_position(('data', 0))
ax2.spines['bottom'].set_position(('data', 0))
ax2.spines[['top', 'right']].set_visible(False)
```
