```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 创建3D坐标系的两种方式对比图
fig = plt.figure(figsize=(12, 5))

# 方式1: Axes3D()
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter([0], [0], [0], c='r', s=100)
ax1.set_xlabel('X轴'), ax1.set_ylabel('Y轴'), ax1.set_zlabel('Z轴')

# 方式2: add_subplot()
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(np.array([[0,1],[0,1]]), np.array([[0,0],[1,1]]), np.array([[0,1],[0,1]]))
plt.tight_layout()
```

```python
# 3D线框图 vs 曲面图
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_wireframe(X, Y, Z, rstride=2, cstride=2)
ax1.set_title('3D线框图')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='coolwarm')
ax2.set_title('3D曲面图')
```

```python
from matplotlib.animation import FuncAnimation

# 正弦波动画帧序列示意图
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10))
    return line,

ani = FuncAnimation(fig, animate, frames=10, interval=200)
```

```python
from mpl_toolkits.basemap import Basemap

# 美国地图示例（简化为轮廓）
fig = plt.figure(figsize=(10, 6))
m = Basemap(projection='merc', llcrnrlat=24, urcrnrlat=50,
            llcrnrlon=-130, urcrnrlon=-65, resolution='l')
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='gray')
```
