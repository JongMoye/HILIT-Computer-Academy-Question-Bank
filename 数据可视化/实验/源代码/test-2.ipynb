```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 折线图
x1 = [1, 3, 4]
y1 = [4, 5, 6]
x2 = [1, 3, 3]
y2 = [7, 8, 9]
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
# 柱形图

x = np.arange(5)
y1 = np.array([10, 8, 7, 11, 13])
y2 = np.array([9, 6, 5, 10, 12])


bar_width = 0.3


plt.figure(figsize=(8, 5))
plt.bar(x, y1,
        tick_label=['a', 'b', 'c', 'd', 'e'],
        width=bar_width,
        label='Group 1',
        color='skyblue')

plt.bar(x + bar_width, y2,
        width=bar_width,
        label='Group 2',
        color='salmon')

plt.legend()
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
x = [0, 1, 2, 3, 4]
y1 = [10, 8, 7, 11, 13]
error = [2, 1, 2.5, 2, 1.5]
bar_width = 0.5

plt.bar(x, y1,
        tick_label=['a', 'b', 'c', 'd', 'e'],
        width=bar_width,
        yerr=error,
        capsize=5,
        color='skyblue',
        edgecolor='black')

plt.title('Bar Chart with Error Bars')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

```python
y = np.arange(5)
x1 = np.array([10, 8, 7, 11, 13])
x2 = np.array([9, 6, 5, 10, 12])


bar_height = 0.3


plt.figure(figsize=(8, 5))
plt.barh(y, x1,
         tick_label=['a', 'b', 'c', 'd', 'e'],
         height=bar_height,
         label='Group 1',
         color='skyblue')

plt.barh(y + bar_height, x2,
         height=bar_height,
         label='Group 2',
         color='salmon')

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
y = [0, 1, 2, 3, 4]
x1 = [10, 8, 7, 11, 13]
x2 = [5, 6, 4, 8, 7]
bar_height = 0.5

plt.barh(y, x1, tick_label=['a', 'b', 'c', 'd', 'e'], height=bar_height)
plt.barh(y, x2, left=x1, height=bar_height)
plt.show()
```

```python
y = [0, 1, 2, 3, 4]
x1 = [10, 8, 7, 11, 13]
x2 = [5, 6, 4, 8, 7]
error = [2, 1, 2.5, 2, 1.5]
bar_height = 0.5

plt.barh(y, x1, tick_label=['a', 'b', 'c', 'd', 'e'], height=bar_height)
plt.barh(y, x2, left=x1, height=bar_height, xerr=error, capsize=5)
plt.show()
```

```python
x = np.arange(6)
y1 = np.array([1, 4, 3, 5, 6, 7])
y2 = np.array([1, 3, 4, 2, 7, 6])
y3 = np.array([3, 4, 3, 6, 5, 5])

plt.stackplot(x, y1, y2, y3)
plt.show()
```
