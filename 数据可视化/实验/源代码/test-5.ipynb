```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
for i in range(1,7):
    plt.subplot(3,2,i)
    plt.text(0.5,0.5, f'Subplot {i}', ha='center', va='center')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
# 顶部跨两列的大子图
ax1 = plt.subplot2grid((3,2), (0,0), colspan=2)
ax1.text(0.5,0.5, 'Growth Rate (2017-2018)', ha='center', va='center')
# 底部两个饼图
ax2 = plt.subplot2grid((3,2), (2,0))
ax2.text(0.5,0.5, '2017 Region', ha='center')
ax3 = plt.subplot2grid((3,2), (2,1))
ax3.text(0.5,0.5, '2018 Region', ha='center')
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,4))
ax1.plot([1,2,3], [4,5,6], 'r-')
ax2.plot([1,2,3], [6,5,4], 'b--')
ax1.set_title('Shared Y-axis')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10,6))
gs = gridspec.GridSpec(2,2)
ax1 = fig.add_subplot(gs[0,:])  # 顶部跨两列
ax2 = fig.add_subplot(gs[1,0])  # 左下
ax3 = fig.add_subplot(gs[1,1])  # 右下

ax1.text(0.5,0.5, 'Sales Overview', ha='center')
ax2.text(0.5,0.5, 'Line Chart', ha='center')
ax3.text(0.5,0.5, 'Area Chart', ha='center')
plt.tight_layout()
plt.show()
```

```python
# 折线图（上方）
plt.subplot(2,1,1)
plt.plot([1,2,3,4], [30,50,45,60], label='Product A')
plt.plot([1,2,3,4], [20,40,35,50], label='Product B')
plt.title('Sales Trend')

# 饼图（下方）
plt.subplot(2,2,3)
plt.pie([60,40], labels=['A','B'], autopct='%1.1f%%')
plt.title('Product A Ratio')

plt.subplot(2,2,4)
plt.pie([50,50], labels=['B','Others'], autopct='%1.1f%%')
plt.title('Product B Ratio')
plt.tight_layout()
plt.show()
```
