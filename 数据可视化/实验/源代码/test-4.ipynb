```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 数据准备
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [23, 45, 32, 56]
product_b = [34, 30, 43, 39]
product_c = [28, 25, 35, 30]

# 创建画布和坐标系
plt.figure(figsize=(10, 6), facecolor='#F5F5F5')

# 绘制三条折线（不同线型+不同标记）
plt.plot(
    quarters, product_a,
    marker='*',                 # 五角星标记
    markersize=10,              # 标记大小
    mfc='gold',                 # 填充色
    mec='navy',                 # 边框色
    mew=1.5,                    # 边框宽度
    linestyle='-',              # 实线
    linewidth=2,
    color='navy',
    label='产品 A'
)

plt.plot(
    quarters, product_b,
    marker='^',                 # 正三角形标记
    markersize=9,
    mfc='#FF4500',
    mec='darkred',
    mew=1.5,
    linestyle='--',             # 短虚线
    linewidth=2,
    color='darkred',
    label='产品 B'
)

plt.plot(
    quarters, product_c,
    marker='s',                 # 正方形标记
    markersize=8,
    mfc='#32CD32',
    mec='darkgreen',
    mew=1.5,
    linestyle='-.',             # 点划线
    linewidth=2,
    color='darkgreen',
    label='产品 C'
)

# 添加图表元素
plt.title('季度销量表', fontsize=14, pad=20)
plt.xlabel('季度', fontsize=12)
plt.ylabel('销量 （百万）', fontsize=12)
plt.grid(linestyle=':', alpha=0.7)

# 自定义刻度标签
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# 添加图例并调整位置
plt.legend(
    loc='upper left',
    frameon=True,
    shadow=True,
    edgecolor='gray',
    fontsize=10,
    title='Product Lines',
    title_fontsize=11
)

# 调整布局
plt.tight_layout()
plt.show()
```
