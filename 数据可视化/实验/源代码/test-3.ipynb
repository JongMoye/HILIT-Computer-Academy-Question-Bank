```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus'] = False
```

```python

# 生成数据
x = np.linspace(-np.pi, np.pi, 256)
y1 = np.sin(x)
y2 = np.cos(x)


# 创建图表和坐标轴
fig, ax = plt.subplots(figsize=(10, 6))

# 绘制曲线
line1, = ax.plot(x, y1, color='blue', linestyle='-', label='正弦曲线')
line2, = ax.plot(x, y2, color='red', linestyle='--', label='余弦曲线')

# 3.2 设置坐标轴
ax.set_xlabel("角度（弧度）", fontsize=12, labelpad=10)
ax.set_ylabel("函数值", fontsize=12, labelpad=10)
ax.set_xlim(-np.pi*1.1, np.pi*1.1)
ax.set_ylim(-1.2, 1.2)
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])

# 3.3 添加标题和图例
ax.set_title("三角函数曲线示例", fontsize=14, pad=20)
ax.legend(loc='upper right', shadow=True, fancybox=True, framealpha=0.8)

# 3.4 显示网格
ax.grid(True, axis='both', linestyle=':', linewidth=0.5, alpha=0.7)

# 3.5 添加参考线和参考区域
ax.axhline(0, color='black', linewidth=0.8)  # x轴参考线
ax.axvline(0, color='black', linewidth=0.8)  # y轴参考线
ax.axhspan(-0.5, 0.5, facecolor='gray', alpha=0.2)  # 水平参考区域

# 3.6 添加注释文本
ax.annotate('最大值点', xy=(np.pi/2, 1), xytext=(np.pi/2+0.5, 0.8),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3'))
ax.text(0.5, -1.1, '基础波形示例', fontsize=10,
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 3.7 添加表格
table_data = [
    ['峰值', '1.0', '-1.0'],
    ['周期', '2π', '2π']
]
ax.table(cellText=table_data,
         colLabels=['特征', '正弦曲线', '余弦曲线'],
         colColours=['#F0F0F0']*3,
         cellLoc='center',
         loc='lower left',
         bbox=[0.02, 0.75, 0.3, 0.2])

# 调整布局
plt.tight_layout()
plt.show()
```
